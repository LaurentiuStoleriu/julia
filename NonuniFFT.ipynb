```julia
using CSV, Tables
M = CSV.read("table.txt", Tables.matrix, header=1);
```

```julia
size(M)[1]
```

```julia
nj = size(M)[1]-1
x = M[2:size(M)[1], 1]
c = M[2:size(M)[1], 6] + 1im * zeros(nj)
```

```julia
size(x)
```

```julia
using Plots
plot(x, abs.(c))
```
