```julia
using Plots
```

```julia
using CSV, Tables
M = CSV.read("table.txt", Tables.matrix, header=1);
```

```julia
using FINUFFT


nj = size(M)[1]-1
x = M[2:size(M)[1], 1];      
c = M[2:size(M)[1], 6] + 1im * zeros(nj)  

n = 501          # output size (number of Fourier modes per dim)
tol = 1e-9        # requested relative tolerance

frqs = Array{Float64}(undef, n);
out = Array{ComplexF64}(undef, n);     # prealloc

nufft1d3!(x, c, 1, tol, frqs, out);         # warm up / compile
```

```julia
time_domain = plot(x, abs.(c))
freq_domain = plot(frqs, abs.(out))
plot(time_domain, freq_domain, layout=2)
```
