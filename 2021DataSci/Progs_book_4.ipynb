```julia
using CairoMakie
CairoMakie.activate!()
```

```julia
fig = scatterlines(1:5, 1:5)
```

```julia
lines(1:10, (1:10).^2; 
        label="x²", 
        linewidth=2, 
        linestyle=nothing,
        figure=(; figure_padding=5, resolution=(600, 400), 
            font="sans", backgroundcolor=:grey90, fontsize=16),
        axis=(; xlabel="x", title="title", xgridstyle=:dash,
            ygridstyle=:dash))
scatterlines!(1:10, (10:-1:1).^2; label="Reverse(x)²")
axislegend("legend"; position=:ct)
current_figure()
```

```julia
using Random: seed!
seed!(28)
xyvals = randn(100, 3)
xyvals[1:5, :]
```

```julia
fig, ax, pltobj = scatter(xyvals[:, 1], xyvals[:, 2]; color=xyvals[:, 3],
label="Bubbles", colormap=:plasma, markersize=15 * abs.(xyvals[:, 3]),
figure=(; resolution=(600, 400)), axis=(; aspect=DataAspect()))
limits!(-3, 3, -3, 3)
Legend(fig[1, 2], ax, valign=:top)
Colorbar(fig[1, 2], pltobj, height=Relative(3 / 4))
fig
caption = "Bubble plot."
current_figure()
```
